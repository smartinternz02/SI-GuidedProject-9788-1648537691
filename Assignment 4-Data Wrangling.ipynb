```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data = pd.read_csv("bank.csv")
```

```python
le = LabelEncoder()
data["job"] = le.fit_transform(data["job"])
data["marital"] = le.fit_transform(data["marital"])
data["education"] = le.fit_transform(data["education"])
data["default"] = le.fit_transform(data["default"])
data["housing"] = le.fit_transform(data["housing"])
data["loan"] = le.fit_transform(data["loan"])
data["contact"] = le.fit_transform(data["contact"])
data["month"] = le.fit_transform(data["month"])
data["poutcome"] = le.fit_transform(data["poutcome"])
data["deposit"] = le.fit_transform(data["deposit"])
```

```python
x = data.iloc[:, 0:16].values
y = data.iloc[:, 16:17].values
print(x.shape)
print(y.shape)
```

```python
x = np.delete(x, [1,2,3,8,10,15], axis=1)
x.shape
```

```python
ohe = OneHotEncoder()
z1 = ohe.fit_transform(data.iloc[:,1:2]).toarray()
z2 = ohe.fit_transform(data.iloc[:,2:3]).toarray()
z3 = ohe.fit_transform(data.iloc[:,3:4]).toarray()
z4 = ohe.fit_transform(data.iloc[:,8:9]).toarray()
z5 = ohe.fit_transform(data.iloc[:,10:11]).toarray()
z6 = ohe.fit_transform(data.iloc[:,15:16]).toarray()
```

```python
x = np.concatenate((z1,z2,z3,z4,z5,z6,x), axis=1)
x.shape
```
